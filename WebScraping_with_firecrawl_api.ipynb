```python
import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r'http[s]?://\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove navigation-related words (customize as needed)
    nav_words = ["menu", "subscribe", "sign up", "footer", "copyright", "privacy policy", "terms of service"]
    pattern = r'\b(?:' + '|'.join(nav_words) + r')\b'
    text = re.sub(pattern, '', text, flags=re.IGNORECASE)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Apply cleaning function to extracted data
cleaned_data = clean_text(d)

# Print cleaned text
print(cleaned_data)
print(cleaned_data[:6000])
sex=cleaned_data[:6000]
```
